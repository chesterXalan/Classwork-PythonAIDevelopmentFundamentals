```python
import numpy as np
import pandas as pd

tips = pd.read_csv('tips.csv')
tips.columns = ['總票價', '小費', '吸菸者', '日期', '時間', '大小']
tips
```

```python
tips['小費比例'] = tips['小費'] / tips['總票價']
tips
```

```python
def top(df: pd.DataFrame, n: int = 6, column: str = '小費') -> pd.DataFrame:
    return df.sort_values(by=column, ascending=False)[:n]
top(tips, n=10)
```

```python
tips.groupby(by='吸菸者').apply(top, n=5)
```
