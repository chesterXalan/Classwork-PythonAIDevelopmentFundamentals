```python
import numpy as np
import pandas as pd

array_2d = np.random.randint(50, 100, size=(50, 5))
df1 = pd.DataFrame(array_2d,
             columns=['國文', '英文', '數學', '地理', '社會'],
             index=[f'{num}號' for num in range(1, 51)])
df1
```

```python
total_s = df1['國文'] + df1['英文'] + df1['數學'] + df1['地理'] + df1['社會']
```

```python
average_s = total_s / 5
average_s
```

```python
df1['總分'] = total_s
df1['平均'] = average_s
df1
```

```python
# Series使用比較運算子
df1['總分'] >= 400
```

```python
# DataFrame使用subscript，搜尋資料
# DataFrame[bool series]
df1[df1['總分'] >= 400]
```

```python
# 查詢國文95分以上的學生
df1[df1['國文'] >= 95]
```

```python
# 查詢國文95分以上的學生同時總分400分以上
# and = &
# or = |
df1[(df1['國文'] >= 95) & (df1['總分'] >= 400)]
```
